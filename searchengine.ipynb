```python
import spacy
import pandas as pd
from collections import defaultdict
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df.drop(['homepage', 'tagline', 'original_language', 'vote_count', 'genres', 'budget', 'id', 'keywords', 'production_companies', 'popularity', 'production_countries', 'spoken_languages', 'revenue', 'original_title'], axis=1, inplace=True)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
def build_inverted_index(df):
  inverted_index = defaultdict(list)

  for idx, row in df.iterrows():
    docs = nlp(row['title'])
    tokens = [token.text for token in docs if not token.is_stop and not token.is_punct]
    title = " ".join(tokens)
    terms = str(title).lower().split()

    for term in terms:
      if idx not in inverted_index[term]:
        inverted_index[term].append(idx)

  return inverted_index
```

```python
inverted_index = build_inverted_index(df)
```

```python
def search_with_inverted_index(query, inverted_index, df, max_results = 5):
  query_terms = query.lower().split()
  matching_docs = set()

  for term in query_terms:
    if term in inverted_index:
      matching_docs.update(inverted_index[term])

  matching_docs = list(matching_docs)[:max_results]

  if matching_docs:
    print(f"\nDisplaying upto {max_results} results for search term {query}: ")
    for doc in matching_docs:
      row = df.iloc[doc]
      print(f"\nMovie title is: {row['title']}")
      print(f"Overview of movie is: {row['overview']}")
      print(f"Status is: {row['status']}")
      print(f"Release date: {row['release_date']}")
      print(f"Its runtime is: {row['runtime']}")
      print(f"Its Rating is: {row['vote_average']}")
      print("-" * 50)
  else:
    print(f"\nNo results found for movie {query}")
```

```python
user_query = input("What movie/series would you like to know about: ")
search_with_inverted_index(user_query, inverted_index, df)
```
